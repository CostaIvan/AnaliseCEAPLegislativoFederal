```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pandas.tools.plotting import table
from matplotlib.lines import Line2D
%matplotlib inline 
sns.set(style="darkgrid")
massa = pd.read_csv('legislativo', delimiter=";", encoding='UTF-8',header=0,index_col = 0)
```

```python
plt.rcParams['figure.figsize'] = (14,8)
```

```python
#Gr√°fico 17 e 18
#Sub 3-3-1

#Selection o Partido a ser analizado
massa_partido = massa[massa.PARTIDO == 'PT']
#Selecionar a categoria dentro do partido
massa_partido_cat = massa_partido[(massa_partido.CAT_DESPESA == 6)] 

#Selecionar Cargos
massa_dep = massa_partido_cat[(massa_partido_cat.CARGO == 'D')].sort_values('VALOR', ascending=False) 
massa_sen = massa_partido_cat[(massa_partido_cat.CARGO == 'S')].sort_values('VALOR', ascending=False) 

#Agrupando os gastos por pessoa
top10_dep = massa_dep.groupby("NOME", as_index=False).sum()
top10_sen = massa_sen.groupby("NOME", as_index=False).sum()

#Organiza os Gastos do Maior para o Menor
top10_dep = top10_dep.sort_values('VALOR',ascending=False)
top10_sen = top10_sen.sort_values('VALOR',ascending=False)

#Plota os caras com os 10 maiores gastos naquela determinada categoria
ax9 = top10_dep.head(10).plot(kind='bar',y='VALOR',x='NOME',color='#DB7B48')
ax9.set_ylabel("Valor em Reais R$")
ax9.set_xlabel("Deputados")
ax9.legend(['Categoria 6 - Passagens a√©reas, aqu√°ticas e terrestres nacionais'])
ax9.set_title('Top 10 Pol√≠ticos com o maior montante reembolsado em um categoria')

ax10 = top10_sen.head(10).plot(kind='bar',y='VALOR',x='NOME',color='b')
ax10.set_ylabel("Valor em Reais R$")
ax10.set_xlabel("Senadores")
ax10.legend(['Categoria 6 - Passagens a√©reas, aqu√°ticas e terrestres nacionais'])
ax10.set_title('Top 10 Pol√≠ticos com o maior montante reembolsado em um categoria')
```

```python
#Gr√°fico 21
#Sub 3-3-3

#Plotando Deputado

#Selection o Partido a ser analizado
massa_partido = massa[massa.PARTIDO == 'PT']
#Selecionar a categoria dentro do partido
massa_partido_cat = massa_partido[(massa_partido.CAT_DESPESA == 6)] 
#Selecionar Cargo - deputado no caso
massa_dep = massa_partido_cat[(massa_partido_cat.CARGO == 'D')].sort_values('VALOR', ascending=False) 

#Agrupar massa deputados
meses_massa_dep = massa_dep.groupby(['MES','NOME'],as_index=False)['VALOR'].sum()

#Tirar M√©dia do per capta do partido
partido_media_percapta_dep = meses_massa_dep.groupby('MES',as_index=False)['VALOR'].mean()


#Selecionar o cara especifico
deputado = massa_dep.loc[massa_dep.NOME == 'Z√â GERALDO']
#Agrupar meses do pol√≠tico
meses_dep = deputado.groupby(['CAT_DESPESA','MES'],as_index=False)['VALOR'].sum()


#plotando Media
ax19 = sns.boxplot(y='VALOR', data=partido_media_percapta_dep,width=0.2,palette=['g'],showfliers=True)
ax19 = sns.swarmplot(y='VALOR',x='MES',data=partido_media_percapta_dep,palette=['g'],size=8, linewidth=1)
#Plotando Politico
ax19 = sns.boxplot(y='VALOR', data=meses_dep,width=0.2,palette=['b'],showfliers=True)
ax19 = sns.swarmplot(y='VALOR',x='MES',data=meses_dep,palette=['b'],size=8, linewidth=1)

bolinha_leg = [Line2D([0],[0],marker='o',color='g'),Line2D([0],[0],marker='o',color='b')];

ax19.set_title('Comparativo mensal de gastos em uma categoria entre os parlamentares de mesmo cargo no mesmo partido')
ax19.legend(bolinha_leg,['Media Deputados do partido','Deputado'])
ax19.set_ylabel('Valor em Reais R$')
ax19.grid(True)
```
