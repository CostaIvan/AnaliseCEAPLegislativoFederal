```python
#CEAPS SENADORES
#Dropa as colunas documento desnecessária
arq_ceap.drop(['DOCUMENTO','DETALHAMENTO'],axis=1, inplace=True)

#TRATANDO COLUNAS

#troca , por . pra poder fazer o typecast da coluna do valor reembolsado
arq_ceap['VALOR_REEMBOLSADO']=arq_ceap['VALOR_REEMBOLSADO'].str.replace(',','.')
#muda o tipo para float
arq_ceap['VALOR_REEMBOLSADO']=arq_ceap['VALOR_REEMBOLSADO'].astype(float)
#valor absoluto para o valor
arq_ceap['VALOR_REEMBOLSADO']=arq_ceap['VALOR_REEMBOLSADO'].abs()

#passar DATA para o formato DATA
#coerce ele tira as datas poha loka tipo a 1017 que ta aparencendo e cola Nat no lugar
arq_ceap['DATA']=pd.to_datetime(arq_ceap['DATA'],format= '%d/%m/%Y', errors ='coerce')
#arq_ceap['DATA']=fillna(0, inplace=True) # Se quiser colocar algum valor no lugar do Nan


#adicionando a coluna CARGO pra marcar que ele é senador
arq_ceap.insert(loc=0, column='CARGO', value='S')

#Renomeando as colunas 
arq_ceap.rename(columns={'SENADOR':'NOME','VALOR_REEMBOLSADO':'VALOR'}, inplace=True)
```

```python
#Faz o left join entre as planilhas
result = arq_ceap.merge(arq_legs,on='NOME',how='left')

#Inserindo coluna CAT_DESPESA
result.insert(loc=0, column='CAT_DESPESA', value='0')

#REORDENANDO
result = result[['ESTADO','PARTIDO','CARGO','NOME','VALOR','TIPO_DESPESA','CNPJ_CPF','FORNECEDOR','CAT_DESPESA','MES','ANO','DATA']]


# Função para classificar as depesas dos senadores

def classifica_despesa_sen (row):
    if row['TIPO_DESPESA'] == 'Aluguel de imóveis para escritório político, compreendendo despesas concernentes a eles.' :
        return '1'
    if row['TIPO_DESPESA'] == 'Aquisição de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos. ' :
        return '2'
    if row['TIPO_DESPESA'] == 'Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar' :
        return '3'
    if row['TIPO_DESPESA'] == 'Divulgação da atividade parlamentar' :
        return '4'
    if row['TIPO_DESPESA'] == 'Locomoção, hospedagem, alimentação, combustíveis e lubrificantes' : 
        return '5'
    if row['TIPO_DESPESA'] == 'Passagens aéreas, aquáticas e terrestres nacionais' :
        return '6'
    if row['TIPO_DESPESA'] == 'Serviços de Segurança Privada' :
        return '7'
    return '0'

#aplicar a função em todas as linhas
result['CAT_DESPESA'] = result.apply (lambda row: classifica_despesa_sen(row),axis=1)
```
